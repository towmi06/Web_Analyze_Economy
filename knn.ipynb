```python
import numpy as np
from sklearn.preprocessing import StandardScaler, KBinsDiscretizer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import pandas as pd

# Làm sạch dữ liệu = thay thế || loại bỏ các giá trị chuỗi có vấn đề
df.replace(' -   ', np.nan, inplace=True)
df.dropna(inplace=True)

# Loại bỏ các khoảng trắng thừa trong tên cột
df.columns = df.columns.str.strip()

target = 'Rate of users Internet (%)'       
# Định nghĩa các nhóm đặc trưng
features1 = ['Mobile device usage (%)', 'CREDIT and BANKING digital transformation ratio(%)', 'Economic Growth Rate of Education (% GDP)']
features2 = ['Revenue of ecommerce platforms (billion USD)', 'Revenue of IT industry (billion USD)', 'Funding in digital infrastructure (billion USD)']
features3 = ['Foreign investment capital (billion USD)']
features4 = ['Number of Enterprises Applying Digital Transformation (bussiness)']

# Định nghĩa các đặc trưng và mục tiêu
X_percent = df[features1]
X_billion_usd = df[features2]
X_billion_usd = df[features3]
X_other = df[features4]
y = df[target]

# Phân loại biến mục tiêu liên tục thành 3 nhóm
discretizer = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='uniform')
y_binned = discretizer.fit_transform(y.values.reshape(-1, 1)).ravel()

# Kết hợp tất cả các nhóm đặc trưng thành một DataFrame duy nhất
X_combined = pd.concat([X_percent, X_billion_usd, X_billion_usd, X_other], axis=1)

# Chia dữ liệu thành tập huấn luyện và kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X_combined, y_binned, test_size=0.25, random_state=45)

# Chuẩn hóa các đặc trưng
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Khởi tạo mô hình KNN
knn = KNeighborsClassifier(n_neighbors=5)

# Huấn luyện mô hình
knn.fit(X_train, y_train)

# Dự đoán
y_pred = knn.predict(X_test)

# Đánh giá mô hình
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
from sklearn.metrics import confusion_matrix, classification_report

# Tính toán ma trận nhầm lẫn
conf_matrix = confusion_matrix(y_test, y_pred)

# Hiển thị Ma trận nhầm lẫn
print("Confusion Matrix:")
print(conf_matrix)

# Hiển thị báo cáo phân loại
class_report = classification_report(y_test, y_pred)
print("\nClassification Report:")
print(class_report)
```

```python
# Vẽ biểu đồ ma trận nhầm lẫn
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Low', 'Medium', 'High'], yticklabels=['Low', 'Medium', 'High'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Ensure all features are numeric
df[features1 + features2 + features3 + features4] = df[features1 + features2 + features3 + features4].apply(pd.to_numeric, errors='coerce')

# Vẽ biểu đồ mật độ cho từng feature theo các nhóm của mục tiêu
plt.figure(figsize=(15, 10))

# Vẽ biểu đồ mật độ cho từng feature trong nhóm X_percent
for feature in features1:
    sns.kdeplot(data=df, x=feature, hue=y_binned, fill=True, common_norm=False)
    plt.title(f'Density Plot for {feature}')
    plt.xlabel(feature)
    plt.ylabel('Density')
    plt.show()

# Vẽ biểu đồ mật độ cho từng feature trong nhóm X_billion_usd
for feature in features2 + features3:
    sns.kdeplot(data=df, x=feature, hue=y_binned, fill=True, common_norm=False)
    plt.title(f'Density Plot for {feature}')
    plt.xlabel(feature)
    plt.ylabel('Density')
    plt.show()

# Vẽ biểu đồ mật độ cho feature trong nhóm X_other
for feature in features4:
    sns.kdeplot(data=df, x=feature, hue=y_binned, fill=True, common_norm=False)
    plt.title(f'Density Plot for {feature}')
    plt.xlabel(feature)
    plt.ylabel('Density')
    plt.show()
```
