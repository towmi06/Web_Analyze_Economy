```python
# khởi tạo feature và target
target = ['CREDIT and BANKING digital transformation ratio(%)', 'Revenue of e-commerce platforms (billion USD)']
features = ['Rate of users Internet (%)', 'Smart Device  (%)', 'Revenue of IT industry (billion USD)', 'Funding in digital infrastructure (billion USD)', 'Foreign investment capital (billion USD)', 'Number of Enterprises Applying Digital Transformation (bussiness)']
X = df1[features]
y = df1[target]
```

### trainning RANDOM FOREST

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

# Define a larger parameter grid for GridSearchCV
param_grid = {
    'n_estimators': [100, 200, 300, 400],  # Added 400 to increase candidates
    'max_features': ['sqrt', 'log2'],
    'max_depth': [75, 100, 125,200],  # Added 125 to increase candidates
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Create a Random Forest Regressor model
rf_model = RandomForestRegressor(random_state=42)

# Use GridSearchCV to find the best hyperparameters
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(X_scaled, y)  # Use the full dataset

# Get the best parameters from GridSearchCV
best_params = grid_search.best_params_
print(f"Best parameters found: {best_params}")

# Train the model with the best parameters on the full dataset
best_rf_model = RandomForestRegressor(**best_params, random_state=42)
best_rf_model.fit(X_scaled, y)
```

```python
# Get the best parameters from GridSearchCV
best_params = grid_search.best_params_
print(f"Best parameters found: {best_params}")

# Reshape X_train to 2D
X_train_reshaped = X_train.reshape(X_train.shape[0], -1)

# Train the model with the best parameters
best_rf_model = RandomForestRegressor(**best_params, random_state=42)
best_rf_model.fit(X_train_reshaped, y_train)
```

```python
# Predict on the test set
y_predict_full = best_rf_model.predict(X_test)
```

```python
# Calculate R^2 score and Mean Squared Error
from sklearn.metrics import r2_score, mean_squared_error

rf_r2 = r2_score(y_test, y_predict_full)
rf_mse = mean_squared_error(y_test, y_predict_full)
print(f'R^2 Score của mô hình Random Forest: {rf_r2}')
print(f'Mean Squared Error của mô hình Random Forest: {rf_mse}')
```

# Training LSTM

```python
from sklearn.preprocessing import MinMaxScaler
%pip install tensorflow
from tensorflow.keras.models import Sequential

from tensorflow.keras.layers import LSTM, Dense
```

```python
# Chọn các cột cần thiết
columns = [
    'Rate of users Internet (%)', 'Smart Device  (%)', 'Revenue of e-commerce platforms (billion USD)',
    'Revenue of IT industry (billion USD)', 'Funding in digital infrastructure (billion USD)',
    'Foreign investment capital (billion USD)', 'CREDIT and BANKING digital transformation ratio(%)',
    'Number of Enterprises Applying Digital Transformation (bussiness)', 'Economic Growth Rate of Education (% GDP)',
    'Number of technology start up businesses'
]
data_filtered = df1[columns]
```

```python
# Chuẩn hóa dữ liệu
scaler = MinMaxScaler()
data_normalized = scaler.fit_transform(data_filtered)
```

```python
# Chia dữ liệu thành tập huấn luyện và kiểm tra
scaler = MinMaxScaler()
data_normalized = scaler.fit_transform(data_filtered)
```

```python

# Tạo dữ liệu đầu vào cho LSTM
def create_dataset(data, time_step=1):
    X, Y = [], []
    for i in range(len(data) - time_step - 1):
        a = data[i:(i + time_step), :]
        X.append(a)
        Y.append(data[i + time_step, 0])
    return np.array(X), np.array(Y)
time_step = 10
X, Y = create_dataset(data_normalized, time_step)
```

```python

# Định hình lại dữ liệu đầu vào cho LSTM
X = X.reshape(X.shape[0], X.shape[1], X.shape[2])
```

```python

# Xây dựng và huấn luyện mô hình LSTM 
model = Sequential()
model.add(LSTM(100, return_sequences=True, input_shape=(time_step, X.shape[2])))
model.add(LSTM(100, return_sequences=False))
model.add(Dense(50))
model.add(Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(X, Y, epochs=200, batch_size=16, verbose=2)
```

```python

# Dự đoán trên tập kiểm tra và tính toán R² score
y_lstm_pred = model.predict(X)
lstm_r2 = r2_score(Y, y_lstm_pred)

print(f'R² Score: {lstm_r2}')
```

```python
# Xuất dữ liệu dự đoán ra file CSV
predictions_df = pd.DataFrame(y_lstm_pred, columns=['Predicted'])
predictions_df.to_csv('predictions.csv', index=False)

print(f'R² Score: {lstm_r2}')
print("Dữ liệu dự đoán đã được lưu vào file predictions.csv")
```

```python
from sklearn.metrics import mean_absolute_error

# Ensure y_test and y_lstm_pred have the same length
min_length = min(len(y_test), len(y_lstm_pred))
y_test_trimmed = y_test[:min_length]
y_lstm_pred_trimmed = y_lstm_pred[:min_length]

# Select the relevant column from y_test_trimmed
y_test_trimmed_single = y_test_trimmed.iloc[:, 0]

mse = mean_squared_error(y_test_trimmed_single, y_lstm_pred_trimmed)
mae = mean_absolute_error(y_test_trimmed_single, y_lstm_pred_trimmed)
rmse = np.sqrt(mse)
```

```python
print(f'Mean Squared Error (MSE): {mse}')
print(f'Mean Absolute Error (MAE): {mae}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
```
