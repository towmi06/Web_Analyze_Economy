```python
# Declare feature vector and target variable
X = df.drop(['Country'], axis = 1)
y = df['Country']
```

```python
# split data into training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
#Find important features with Random Forest model

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
# Thêm cột mục tiêu vào tập huấn luyện để dễ lọc
df_train = X_train.copy()
df_train['Target'] = y_train

df_test = X_test.copy()
df_test['Target'] = y_test

# Loại bỏ cột 'year' khỏi DataFrame
df_train = df_train.drop(columns=["Year"], errors='ignore')
df_test = df_test.drop(columns=["Year"], errors='ignore')

# Lặp qua từng châu lục và huấn luyện mô hình riêng
continents = df_train["Continents"].unique()

for continent in continents:
    # Lọc dữ liệu cho từng châu lục trong tập huấn luyện và kiểm tra
    train_continent = df_train[df_train["Continents"] == continent]
    test_continent = df_test[df_test["Continents"] == continent]

    # Đặt X và y cho từng châu lục
    X_train_continent = train_continent.drop(columns=["Target", "Continents"])
    y_train_continent = train_continent["Target"]

    X_test_continent = test_continent.drop(columns=["Target", "Continents"])
    y_test_continent = test_continent["Target"]

    # Khởi tạo và huấn luyện mô hình cho từng châu lục
    rfc = RandomForestClassifier(random_state=0, n_estimators=10)
    rfc.fit(X_train_continent, y_train_continent)

    # Dự đoán trên tập kiểm tra của từng châu lục
    y_pred_continent = rfc.predict(X_test_continent)

    # Tính độ chính xác và hiển thị kết quả
    accuracy = accuracy_score(y_test_continent, y_pred_continent)
    print(f"Model accuracy score for {continent} with 10 decision-trees: {accuracy:.4f}")

    # Lấy độ quan trọng của các đặc trưng
    importances = rfc.feature_importances_
    feature_names = X_train_continent.columns
    feature_importances = pd.DataFrame(importances, index=feature_names, columns=["Importance"]).sort_values("Importance", ascending=False)

    # Hiển thị các đặc trưng quan trọng cho châu lục
    print(f"\nCác đặc trưng quan trọng cho {continent}:\n", feature_importances)

    # # Tạo confusion matrix cho từng châu lục
    # cm = confusion_matrix(y_test_continent, y_pred_continent)

    # # In ra confusion matrix
    # print(f'Confusion matrix for {continent}:\n\n', cm)
```

```python
# Random Forest Classifier model with all parameter n_estimators=100
clf_all_features = RandomForestClassifier(n_estimators=100, random_state=0)
clf_all_features.fit(X_train_continent, y_train_continent)

# Dự đoán trên tập kiểm tra và tính độ chính xác
y_pred_all_features = clf_all_features.predict(X_test_continent)
accuracy_all_features = accuracy_score(y_test_continent, y_pred_all_features)

print(f"Accuracy of the model with all features: {accuracy_all_features:.4f}")
```

```python
#Visualize the feature scores of the features

importance_dict = {}

# Loop importances
continents = df_train["Continents"].unique()

for continent in continents:

    train_continent = df_train[df_train["Continents"] == continent]

    # Define features and target variable
    X_train_continent = train_continent.drop(columns=["Target", "Continents"])
    y_train_continent = train_continent["Target"]

    # Check if there are features available for training
    if X_train_continent.shape[1] > 0:
        # Create and fit the model to the training set
        clf = RandomForestClassifier(n_estimators=100, random_state=0)
        clf.fit(X_train_continent, y_train_continent)

        importance_dict[continent] = clf.feature_importances_

# Chuyển từ dictionary sang DataFrame để vẽ
importance_df = pd.DataFrame(importance_dict, index=X_train_continent.columns)

# Check if the DataFrame is not empty and has the correct shape
if not importance_df.empty:
    # Plotting feature importances for each continent
    plt.figure(figsize=(14, 8))
    importance_df.plot(kind='bar', figsize=(14, 8), alpha=0.7)

    # Add labels to the graph
    plt.xlabel('Features', fontsize=12)
    plt.ylabel('Feature Importance Score', fontsize=12)

    # Add title to the graph
    plt.title("Visualizing Important Features by Continent", fontsize=14)


    plt.xticks(rotation=30, ha='right', fontsize=10)

    # Add a legend for the continents
    plt.legend(title='Continents', fontsize=10)

    # Adjust layout
    plt.tight_layout()

    # Show the plot
    plt.show()
else:
    print("No feature importances to plot.")
```

```python
#fix lại xíu data theo châu lục cơ....
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt


# Lọc danh sách các châu lục, bỏ qua America và Africa
continents_to_display = [continent for continent in continents if continent not in ['America', 'Africa ']]

# Lặp qua từng châu lục và hiển thị confusion matrix
for continent in continents_to_display:
    # Lọc dữ liệu cho từng châu lục trong tập huấn luyện và kiểm tra
    train_continent = df_train[df_train["Continents"] == continent]
    test_continent = df_test[df_test["Continents"] == continent]

    # Đặt X và y cho từng châu lục
    X_train_continent = train_continent.drop(columns=["Target", "Continents"])
    y_train_continent = train_continent["Target"]

    X_test_continent = test_continent.drop(columns=["Target", "Continents"])
    y_test_continent = test_continent["Target"]

    # Khởi tạo và huấn luyện mô hình cho từng châu lục
    rfc = RandomForestClassifier(random_state=0, n_estimators=10)
    rfc.fit(X_train_continent, y_train_continent)

    # Dự đoán trên tập kiểm tra của từng châu lục
    y_pred_continent = rfc.predict(X_test_continent)

    # Tạo confusion matrix cho từng châu lục
    cm = confusion_matrix(y_test_continent, y_pred_continent)

    # In ra confusion matrix
    # print(f'Confusion matrix for {continent}:\n\n', cm)

    # Hiển thị confusion matrix bằng biểu đồ với kích thước tùy chỉnh
    plt.figure(figsize=(10, 7))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap=plt.cm.Blues)

    # Tinh chỉnh các nhãn trục
    plt.xticks(rotation=45, fontsize=10)
    plt.yticks(fontsize=10)

    plt.title(f"Confusion Matrix for {continent}", fontsize=14)
    plt.tight_layout()
    plt.show()
```

```python
from sklearn.metrics import classification_report


for continent in continents:

    train_continent = df_train[df_train["Continents"] == continent]
    test_continent = df_test[df_test["Continents"] == continent]

    # Đặt X và y cho từng châu lục
    X_train_continent = train_continent.drop(columns=["Target", "Continents"])
    y_train_continent = train_continent["Target"]

    X_test_continent = test_continent.drop(columns=["Target", "Continents"])
    y_test_continent = test_continent["Target"]

    # Khởi tạo và huấn luyện mô hình cho từng châu lục
    rfc = RandomForestClassifier(random_state=0, n_estimators=10)
    rfc.fit(X_train_continent, y_train_continent)

    # Dự đoán trên tập kiểm tra của từng châu lục
    y_pred_continent = rfc.predict(X_test_continent)

    # Tạo và in báo cáo phân loại
    report = classification_report(y_test_continent, y_pred_continent)
    print(f"Classification report for {continent}:\n{report}\n")
```
